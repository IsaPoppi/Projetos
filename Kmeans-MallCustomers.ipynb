# 5.1 - Relatório de Classificação

O relatório de classificação fornece uma visão abrangente das medidas de capacidade preditiva do modelo para cada classe (Não Inadimplente e Inadimplente), além de médias ponderadas e macro. Aqui estão os detalhes:

**Precisão:** A razão entre os verdadeiros positivos e todos os positivos previstos. Indica a capacidade do classificador de não rotular uma instância negativa como positiva.

**Recall (Sensibilidade):** A razão entre os verdadeiros positivos e todos os positivos reais. Mede a capacidade do classificador de encontrar todas as instâncias positivas.

**F1-Score:** A média harmônica de precisão e recall. Fornecendo um equilíbrio entre a precisão e o recall.

**Support:** O número de ocorrências reais de cada classe no conjunto de dados especificado.


```python
from sklearn.metrics import classification_report

# Gerando o relatório de classificação
report = classification_report(y_test, y_pred, target_names=['Não Inadimplente', 'Inadimplente'])

print(report)
```

O *F1-score* para **"Não Inadimplente"** é 0.67 e para **"Inadimplente"** é 0.26.

A *acurácia* geral do modelo é 0.54.

Esses resultados sugerem que o modelo é relativamente melhor em prever clientes "Não Inadimplentes" do que "Inadimplentes", como evidenciado pelas maiores pontuações de precisão e F1 para a classe "Não Inadimplente".
